## Crear columnas

```python
datos = pd.read_csv(url, sep = ';')
datos.shape
```

```python
datos['Valor Mensual'] = datos['Valor'] + datos['Condominio']
datos.head()
```

```python
datos['Valor Anual'] = datos['Valor Mensual'] * 12 + datos['Impuesto']
datos.tail()
```

## Crear columnas categoricas

```python
datos['Descripcion'] = datos['Tipo'] + ' en la colonia ' +  datos['Colonia'] + \
                    ' con el numero de habitaciones igual a ' + datos['Habitaciones'].astype(str) + \
                    ' con el numero de garages igual a ' + datos['Garages'].astype(str)
datos.head()
```

## Crear columna para ver si tiene suite

```python
datos['Tiene suite'] = datos['Suites'].apply(lambda x: 'Si' if x > 0 else 'No')
```
